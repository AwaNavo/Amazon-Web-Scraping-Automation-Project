```python
# Getting Data from Soup

product_title = Soup.find(id='productTitle').text.strip()
print(product_title)

price_whole = Soup.find(class_='a-price-whole').text.strip()
print(price_whole)

price_fraction = Soup.find(class_='a-price-fraction').text.strip()
print (price_fraction) 

price = price_whole + price_fraction
print(price)

review_score = Soup.find(class_ = 'a-size-small a-color-base').text.strip()
print(review_score)

total_reviews = Soup.find(id='acrCustomerReviewText').text.strip('()')
print(total_reviews)
```

```python
# in order to automate the process, creating a function - check_price()
def check_price():
    
    global file_path
    global csv_header,csv_data
    
    
    URL = r"https://www.amazon.com/ebossy-Womens-Striped-Button-Pockets/dp/B0FGJGMCMP/ref=sr_1_2?dib=eyJ2IjoiMSJ9.TyJ2OLUjrkmyEb1SQTG5bfdUhAEHaaNiZkg6WJdDceeSZR-pYQMw7gVLZWSXJvAD-JZ1hhZg4u3CwGWXEllb7AYc9KMGWhASkth6Q_mJHh9BPlXuyoMUBt6CYkREUAbThQe-y5lSS2A2YuwJd9HIuM_GOQjC8oq6CvonTDmLV3hG9zRA6yqIfjdtKdC0QRwjccNp9pGHvDmf2MspvRZpxynapYj5-3unbsKAK1JGsRwtCcpfZEP6jw5-gN0dRq9ft0_tUCzUZ3H6Y8WEw3mfLN0lCiNrLkFh-SbhcfTcTPw.sYzOYWqKkOoXYI61TyW5YnG02JeGsU99wb6QQaCJ3I0&dib_tag=se&keywords=linen%2Bdress&qid=1772526705&sr=8-2&th=1&psc=1"
    header = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
        "Accept-Language": "en-US,en;q=0.9",
        "Accept-Encoding": "gzip, deflate, br",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8",
        "Connection": "keep-alive",
    }
    
    page = requests.get(URL,headers = header)
    Soup = BeautifulSoup(page.content,'html.parser')
    # print(Soup.prettify())

    # getting data
    product_title = Soup.find(id='productTitle').text.strip()
    price_whole = Soup.find(class_='a-price-whole').text.strip()
    price_fraction = Soup.find(class_='a-price-fraction').text.strip()
    price = price_whole + price_fraction
    review_score = Soup.find(class_ = 'a-size-small a-color-base').text.strip()
    total_reviews = Soup.find(id='acrCustomerReviewText').text.strip('()')
   
    # csv file definition
    csv_header = ['Product_Title','Price','Review_Score','Total_Reviews','Date']
    csv_data = [product_title,price,review_score,total_reviews,date]

    file_path = r'C:\Users\User\Downloads\Amazon_Web_Scraping_Dataset.csv'

    
    if not os.path.isfile(file_path):
    
        with open(file_path, mode='w', newline='', encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow(csv_header)
            writer.writerow(csv_data)
    else:
            
        with open(file_path, mode='a+', newline='', encoding='UTF8') as f:
            writer = csv.writer(f)
            writer.writerow(csv_data)

    
```

```python
# authomated the updating file process - for required time 
logging.basicConfig(level=logging.INFO)

while(True):
    try:
        logging.info('Checking price...')
        check_price()
        
        print(f'Checked Amazon website - {'https://www.amazon.com/'} at:{datetime.now()}')
        time.sleep(10)  # sleep for 10sec, then run the loop again
        
    except Exception as e:
        logging.error(f'Error occured: {e}')
        time.sleep(5)
    except KeyboardInterrupt:
        logging.info('Process stopped by user.')
        break
```
